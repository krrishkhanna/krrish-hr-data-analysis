```python

# Q1: Variance, Standard Deviation, and IQR of Salaries in HR and Marketing Departments
hr_marketing = df[df['Department'].isin(['HR', 'Marketing'])]

# Variance of salaries
variance_hr = hr_marketing[hr_marketing['Department'] == 'HR']['Salary'].var()
variance_marketing = hr_marketing[hr_marketing['Department'] == 'Marketing']['Salary'].var()

# Standard deviation of salaries
std_dev_hr = hr_marketing[hr_marketing['Department'] == 'HR']['Salary'].std()
std_dev_marketing = hr_marketing[hr_marketing['Department'] == 'Marketing']['Salary'].std()

# IQR of salaries
iqr_hr = hr_marketing[hr_marketing['Department'] == 'HR']['Salary'].quantile(0.75) - hr_marketing[hr_marketing['Department'] == 'HR']['Salary'].quantile(0.25)
iqr_marketing = hr_marketing[hr_marketing['Department'] == 'Marketing']['Salary'].quantile(0.75) - hr_marketing[hr_marketing['Department'] == 'Marketing']['Salary'].quantile(0.25)

# Display results
print(f"HR - Variance: {variance_hr}, Std Dev: {std_dev_hr}, IQR: {iqr_hr}")
print(f"Marketing - Variance: {variance_marketing}, Std Dev: {std_dev_marketing}, IQR: {iqr_marketing}")

# Q2: Range of Experience in IT Department
it_experience = df[df['Department'] == 'IT']['Experience']
experience_range = it_experience.max() - it_experience.min()
print(f"Range of experience in IT department: {experience_range}")

# Q3: Boxplot for Experience in Each Department
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Department', y='Experience', data=df)
plt.title('Experience by Department')
plt.xticks(rotation=90)
plt.show()

# Q4: Feature with the Highest Standard Deviation
std_devs = df.select_dtypes(include='number').std()
highest_std_feature = std_devs.idxmax()
highest_std_value = std_devs.max()

print(f"Feature with highest standard deviation: {highest_std_feature}, Std Dev: {highest_std_value}")
```
